```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\new data\datasets\Walmart_data.csv')
```

```python
# Which months had the highest and lowest sales?

df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.strftime('%b')

#month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 
#               'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# df['Month']=pd.Categorical(df['Month'],categories=month_order, ordered=True)

monthly_sales = df.groupby('Month')['Total'].sum().reset_index()

plt.figure(figsize=(6,5))

sns.lineplot(data=monthly_sales,x='Month',y='Total', marker='o')

plt.title('Monthly Sales Trend')

plt.xlabel('Month')

plt.ylabel('Total Sales')

plt.tight_layout()

plt.show()
```

```python
# Which branch had the highest total sales across the 3 months.

Highest_sales = df.groupby('Branch')['Total'].sum().reset_index()

plt.figure(figsize=(4,3))

sns.barplot(data=Highest_sales,x='Branch',y='Total')

plt.title('Highest monthly sales')

plt.xlabel('Branch')

plt.ylabel('Total Sales')

plt.tight_layout()

plt.show()
```

```python
# Which product category generated the most revenue.

Product_revenue = df.groupby('Product_line')['Total'].sum()

Product_revenue.plot.pie(autopct='%1.1f%%', figsize=(4,4))

plt.title('Highest Revenue Product')

plt.ylabel('')

plt.show()
```

```python
# Which type of customer (Member vs Normal) spends more on average

Customer_spend = df.groupby('Customer_type')['Total'].mean().reset_index()

plt.figure(figsize=(3,3))

sns.barplot(data=Customer_spend,x='Customer_type',y='Total')

plt.title('Customer type spend')

plt.xlabel('Customer type')

plt.ylabel('Avergae spends')

plt.tight_layout()

plt.show()
```

```python
# What is the gender distribution of customers

df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(4,4))

plt.title('Gender based customers')

plt.ylabel('')

plt.show()
```

```python
# Which time of day has the most sales? (Morning, Afternoon, Evening)

df['Time']=pd.to_datetime(df['Time'])

def get_time_of_day(hour):
    if 5<= hour < 12:
        return 'Morning'
    elif 12 <= hour <17:
        return 'Afternoon'
    else:
        return 'Evening'
    
df['Time_of_day']=df['Time'].dt.hour.apply(get_time_of_day)   

Customer_spend = df.groupby('Time_of_day')['Total'].sum().reset_index()

plt.figure(figsize=(4,4))

sns.barplot(data=Customer_spend,x='Time_of_day',y='Total')

plt.title('Time Based Sales')

plt.xlabel('Time of day')

plt.ylabel('Most sales')

plt.tight_layout()

plt.show()    
```

```python
# Which day of the week sees the most sales

df['Date']=pd.to_datetime(df['Date'])

df['Day'] = df['Date'].dt.day_name()

Day_sales = df.groupby('Day')['Total'].sum().reset_index()

# Correct list of week days
Day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

Day_sales['Day']=pd.Categorical(Day_sales['Day'],categories=Day_order, ordered=True)

Day_sales = Day_sales.sort_values('Day')

sns.barplot(data=Day_sales,x='Day',y='Total')

plt.title('Day Based Sales')

plt.xlabel('Week days')

plt.ylabel('Most sales')

plt.tight_layout()

plt.show()   
```
